```python
from pathlib import Path
from dataclasses import dataclass
import time

RAPL_PATH = Path("/sys/class/powercap/intel-rapl:0/energy_uj")
MAX_PATH  = RAPL_PATH.with_name("max_energy_range_uj")

@dataclass
class Sample:
    t_ms: int          # milliseconds
    energy_uj: int     # microjoules

def compute_stats(start: Sample, end: Sample):
    """
    Given two samples, compute:
      - elapsed seconds
      - energy difference in Joules
      - average power in Watts
    """
    dt_s = (end.t_ms - start.t_ms) / 1000.0
    if dt_s <= 0:
        raise ValueError("Non-positive elapsed time between samples.")

    max_uj = 65532610987 # This is defined by "/sys/class/powercap/intel-rapl:0/max_energy_range_uj", lets the value wrap back around after hitting the max
    delta_uj = end.energy_uj - start.energy_uj
    if delta_uj < 0:  # wraparound
        delta_uj += max_uj

    dJ = delta_uj / 1e6 # micro Joules to Joules
    avg_W = dJ / dt_s # Calculate average wattage over time
    print(f"Elapsed: {dt_s:.3f} s  |  Energy: {dJ:.3f} J  |  Avg Power: {avg_W:.2f} W")
```
